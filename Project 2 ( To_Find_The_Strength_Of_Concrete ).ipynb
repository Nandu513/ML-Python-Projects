```python
def remove_outlier(co1):
    sorted(co1)
    Q1,Q3=co1.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5+IQR)
    upper_range=Q3+(1.5+IQR)
    return lower_range,upper_range
```

```python
lowerage,upperage=remove_outlier(concrete_df['age'])
concrete_df['age']=np.where(concrete_df['age']>upperage,upperage,concrete_df['age'])
concrete_df['age']=np.where(concrete_df['age']<lowerage,lowerage,concrete_df['age'])

lowerstrength,upperstrength=remove_outlier(concrete_df['strength'])
concrete_df['strength']=np.where(concrete_df['strength']>upperstrength,upperstrength,concrete_df['strength'])
concrete_df['strength']=np.where(concrete_df['strength']<lowerstrength,lowerstrength,concrete_df['strength'])

lowwater,highwater=remove_outlier(concrete_df['water'])
concrete_df['water']=np.where(concrete_df['water']>highwater,highwater,concrete_df['water'])
concrete_df['water']=np.where(concrete_df['water']<lowwater,lowwater,concrete_df['water'])

lowsuperplastic,highsuperplastic=remove_outlier(concrete_df['superplastic'])
concrete_df['superplastic']=np.where(concrete_df['superplastic']>highsuperplastic,highsuperplastic,concrete_df['superplastic'])
concrete_df['superplastic']=np.where(concrete_df['superplastic']<lowsuperplastic,lowsuperplastic,concrete_df['superplastic'])

lowfineagg,highfineagg=remove_outlier(concrete_df['fineagg'])
concrete_df['fineagg']=np.where(concrete_df['fineagg']>highfineagg,highfineagg,concrete_df['fineagg'])
concrete_df['fineagg']=np.where(concrete_df['fineagg']<lowfineagg,lowfineagg,concrete_df['fineagg'])

lowslag,highslag=remove_outlier(concrete_df['slag'])
concrete_df['slag']=np.where(concrete_df['slag']>highslag,highslag,concrete_df['slag'])
concrete_df['slag']=np.where(concrete_df['slag']<lowslag,lowslag,concrete_df['slag'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = concrete_df.drop(['strength'],axis=1)
y = concrete_df['strength']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=2)
```

```python
reg_model=LinearRegression()
reg_model.fit(x_train,y_train)
```

```python
reg_model.score(x_train,y_train)
```

```python
reg_model.score(x_test,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
```

```python
poly=PolynomialFeatures(degree=2,interaction_only=True)
x_train2=poly.fit_transform(x_train)
x_test2=poly.fit_transform(x_test)
poly_clf=linear_model.LinearRegression()
poly_clf.fit(x_train2,y_train)
print(poly_clf.score(x_train2,y_train))
```

```python
print(poly_clf.score(x_test2,y_test))
```

```python
poly=PolynomialFeatures(degree=1,interaction_only=True)
x_train2=poly.fit_transform(x_train)
x_test2=poly.fit_transform(x_test)
poly_clf=linear_model.LinearRegression()
poly_clf.fit(x_train2,y_train)
print(poly_clf.score(x_train2,y_train))
```

```python
print(poly_clf.score(x_test2,y_test))
```

```python
from sklearn.preprocessing import StandardScaler
stand= StandardScaler()
Fit = stand.fit(x_train)
x_train_scl = Fit.transform(x_train)
x_test_scl = Fit.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr=LinearRegression()
fit=lr.fit(x_train_scl,y_train)
score = lr.score(x_test_scl,y_test)
print('predcted score is : {}'.format(score))
print('..................................')
y_predict = lr.predict(x_test_scl)
print('mean_sqrd_error is ==',mean_squared_error(y_test,y_predict))
rms = np.sqrt(mean_squared_error(y_test,y_predict)) 
print('root mean squared error is == {}'.format(rms))
```

```python
plt.figure(figsize=[16,8])
plt.scatter(y_predict,y_test)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red')
plt.xlabel('predicted')
plt.ylabel('orignal')
plt.show()
```

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import mean_squared_error
rd= Ridge(alpha=0.4)
ls= Lasso(alpha=0.3)
fit_rd=rd.fit(x_train_scl,y_train)
fit_ls = ls.fit(x_train_scl,y_train)
print('score od ridge regression is:-',rd.score(x_test_scl,y_test))
print('.......................................................')
print('score of lasso is:-',ls.score(x_test_scl,y_test))
print('mean_sqrd_roor of ridig is==',mean_squared_error(y_test,rd.predict(x_test_scl)))
print('mean_sqrd_roor of lasso is==',mean_squared_error(y_test,ls.predict(x_test_scl)))
print('root_mean_squared error of ridge is==',np.sqrt(mean_squared_error(y_test,rd.predict(x_test_scl))))
print('root_mean_squared error of lasso is==',np.sqrt(mean_squared_error(y_test,lr.predict(x_test_scl))))
```
