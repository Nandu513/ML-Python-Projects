```python
def remove_outlier(co1):
    sorted(co1)
    Q1,Q3=co1.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5+IQR)
    upper_range=Q3+(1.5+IQR)
    return lower_range,upper_range
```

```python
lowamount,highamount=remove_outlier(credit_df['amount'])
credit_df['amount']=np.where(credit_df['amount']>highamount,highamount,credit_df['amount'])
credit_df['amount']=np.where(credit_df['amount']<lowamount,lowamount,credit_df['amount'])

lowmonths_loan_duration,highmonths_loan_duration=remove_outlier(credit_df['months_loan_duration'])
credit_df['months_loan_duration']=np.where(credit_df['months_loan_duration']>highmonths_loan_duration,highmonths_loan_duration,credit_df['months_loan_duration'])
credit_df['months_loan_durationq']=np.where(credit_df['months_loan_duration']<lowmonths_loan_duration,lowmonths_loan_duration,credit_df['months_loan_duration'])

lowage,highage=remove_outlier(credit_df['age'])
credit_df['age']=np.where(credit_df['age']>highage,highage,credit_df['age'])
credit_df['age']=np.where(credit_df['age']<lowage,lowage,credit_df['age'])
```

### Random Forest classifier 

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
```

```python
x=credit_df[['months_loan_duration', 'amount', 'percent_of_income',
       'years_at_residence', 'age', 'dependents']]
y=credit_df[['existing_loans_count']]
```

```python
scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size=0.2, random_state = 4)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, random_state=0)
rfc.fit(x_train, y_train)
```

```python
feature_imp = pd.Series(rfc.feature_importances_, index=x.columns).sort_values(ascending=False)
feature_imp
```

```python
f, ax = plt.subplots(figsize=(14, 7))
ax = sns.barplot(x=feature_imp, y=feature_imp.index)
ax.set_title("Visualize feature scores of the features")
ax.set_yticklabels(feature_imp.index)
ax.set_xlabel("Feature importance score")
ax.set_ylabel("Features")
plt.show()
```
